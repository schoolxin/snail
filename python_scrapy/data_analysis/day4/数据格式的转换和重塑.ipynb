### 在字段上使用函数方法进行处理，Map, Apply, ApplyMap

```python
#map方法的使用，map方法针对列中的每个元素进行处理
s = '1.5元/平米・月'
s.split('元')[0]
```

```python
def removeDollar(e):
    return e.split('元')[0]
```

```python
df['物 业 费'].map(removeDollar)
```

```python
df['物 业 费'].map(lambda e: e.split('元')[0])
```

```python
#apply方法可以针对整行或整列数据进行操作，axis=0为列， axis=1为行，有点类似于sql聚合函数。

df1 = pd.DataFrame([
    [60, 70, 50],
    [80, 79, 68],
    [63, 66, 82]
], columns=['first', 'second', 'third'])
df1
```

```python
df1.apply(lambda e: e.max() - e.min(), axis=1)
```

```python
#applymap是针对整个dataframe里面的每个元素进行操作
def convertNaN(e):
    if e == '暂无资料':
        return np.nan
    else:
        return e

df.applymap(convertNaN).head()
```

```python
df.applymap(lambda e: np.nan if e=='暂无资料' else e).head()
```

## 时间日期格式的处理

```python
from datetime import datetime, timedelta

s = '2018/12/31'
datetime.strptime(s, '%Y/%m/%d')
```

```python
pd.to_datetime(df['张贴日期'], format='%Y年%m月%d日').head()
```

```python
df['张贴日期'] = pd.to_datetime(df['张贴日期'], format='%Y年%m月%d日')
```

```python
df.head()
```

```python
df['张贴日期'].head()
```

```python
df['张贴日期'].iloc[0] - timedelta(days=1)
```

```python
for i in range(1, 11):
    dt = df['张贴日期'].iloc[0] - timedelta(days=i)
    print(dt)
```
